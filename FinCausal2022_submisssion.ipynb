Here are general scenarios to keep in mind (and avoid) while submitting:

*   Empty "Cause" and/or "Effect" columns
*   Uneven (longer/shorter) submissions
*   Random fillings (i.e., filling the "Cause" and/or "Effect" columns with chunks that do not appear in the corresponding "Text" column)
